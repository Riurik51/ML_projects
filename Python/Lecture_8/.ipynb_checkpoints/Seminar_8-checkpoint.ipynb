```python
class Fibonachi_iterator:
    def __init__(self, n):
        self.n = n
        self.iter = 0
        self.pred = 1
        self.predpred = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.iter < self.n:
            temp = self.predpred
            self.iter +=1
            self.predpred = self.pred
            self.pred += temp    
            return temp
        else:
            raise StopIteration()
```

```python
for i in Fibonachi_iterator(109):
    print(i)
```
