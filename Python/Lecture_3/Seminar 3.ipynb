# Преобразовать скачивания в число

```python
apps_filter = apps[apps['Installs'] != 'Free']
```

```python
apps_filter['Installs'].unique()
```

```python
def to_number(string):
    string = int(string.replace('+','').replace(',',''))
    return string
```

```python
apps_filter.loc[:,'Installs_number'] = apps_filter.loc[:,'Installs'].apply(to_number)
```

```python
apps_filter.head()
```

# Вывести самое рейтинговое приложение в самой скачиваемой категории

```python
category = apps_filter[['Category','Installs_number']].groupby(['Category'])[['Installs_number']].sum()
maxmax = apps_filter[apps_filter['Category'] == category['Installs_number'].idxmax()]
maxmax[maxmax['Rating'] == 5]
```
