__11__. Приведем столбец `TRDATETIME` к типу даты.

```python
from datetime import datetime

df["TRDATETIME"] = pd.to_datetime(df["TRDATETIME"], format='%d%b%y:%H:%M:%S')
```

Извлеките из столбца `TRDATETIME` следующие признаки:
* `mday` – день (с начала месяца);
* `ymonth` – месяц;
* `year` – год;
* `wday` – день недели;
* `is_weekend` – является ли день выходным днем.

Извлеченные признаки запишите в таблицу `df_time`.

```python
def date_to_year(date):
    return int(date.year)


def date_to_day(date):
    return int(date.day)


def date_to_wday(date):
    return int(date.strftime("%w")) - 1


def date_to_ymonth(date):
    return int(date.month)


def date_is_weekend(date):
    if int(date.strftime("%w")) >= 5:
        return True
    else: 
        return False
```

```python
df_time = pd.DataFrame()
df_time['is_weekend'] = df["TRDATETIME"].apply(date_is_weekend)
df_time['mday'] = df["TRDATETIME"].apply(date_to_day)
df_time['wday'] = df["TRDATETIME"].apply(date_to_wday)
df_time['year'] = df["TRDATETIME"].apply(date_to_year)
df_time['ymonth'] = df["TRDATETIME"].apply(date_to_ymonth)
```

```python
df_time.head()
```
