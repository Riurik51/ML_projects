# ДЗ Сентимент-анализ предложений с помощью линейных моделей 
Проничкин Юрий

$$ Q = \frac{1}{l} \sum_{i = 1}^{l} log(1 + e^{-y_i(X_i, w)}) + \frac{1}{2}||w||^2 \dashrightarrow min $$

Значит градиент этой функции:
$$ \nabla Q = \frac{1}{l} \sum_{i = 1}^{l} \frac{-y_iX_i}{1 + e^{y_i(X_i, w)}} + w $$

Здесь X - выборка из $l$ векторов размерности $n$, y - ответы на выборке X, w - вектор весов размерности $n$.

Импортируем нужные библиотеки

```python
import pandas as pd
from optimization import GDClassifier, SGDClassifier
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

Читаем тренировочные данные

```python
df = pd.read_csv('toxic_train.csv')
```

```python
y = np.array(df['is_toxic'], dtype=int)
```

```python
y = 2* y - 1
```

Функция для предобработки - убираем лишнее, приводим к нижнему регистру.

```python
def preprocessing(string):
    string = string.lower()
    string = re.sub('[^\w\s]', ' ', string)
    string = ' '.join(string.split('\n'))
    return string
```

Создаем разряженную матрицу - тренировочный датасет с помощью CountVectorizer

```python
corpus = list(df['comment_text'].map(preprocessing))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(min_df=4)
```

```python
X = vectorizer.fit_transform(corpus)
```

То же самое с тестовыми данными

```python
df = pd.read_csv('toxic_test.csv')
```

```python
y_test = np.array(df['is_toxic'], dtype=int)
y_test = y_test * 2 - 1
```

```python
corpus = list(df['comment_text'].map(preprocessing))
```

```python
X_test = vectorizer.transform(corpus)
```

Добавим константный признак

```python
X = sp.sparse.hstack((X, np.ones(X.shape[0])[:, None]))
```

```python
X_test = sp.sparse.hstack((X_test, np.ones(X_test.shape[0])[:, None]))
```

Проверим сходимость в зависимости от step_alpha, здесь step_beta=0, w_0=0

```python
alpha = np.logspace(-3, 2, 4)
to_drow = {}
```

```python
w_0 = np.zeros(X.shape[1])
```

```python
for step_alpha in alpha:
    gdc = GDClassifier('binary_logistic',
                    step_alpha=step_alpha,
                    step_beta=0,
                    l2_coef=1e-6)
    history = gdc.fit(X,
                  y,
                  w_0,
                  True,
                  X_test=X_test,
                  y_test=y_test)
    to_drow[f'{step_alpha}'] = history
```

```python
plt.figure(figsize=(9,6), dpi=75)
for step_alpha in alpha:
    plt.plot(np.cumsum(to_drow[f'{step_alpha}']['time']), np.log1p(to_drow[f'{step_alpha}']['func']), label=f'{step_alpha}')
plt.xlabel('Время', fontsize=14)          
plt.ylabel('log(1 + Loss)', fontsize=14)
plt.title('Loss от времени', fontsize=22)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,6), dpi=75)
for step_alpha in alpha:
    plt.plot(range(1, 1002), np.log1p(to_drow[f'{step_alpha}']['func']), label=f'{step_alpha}')
plt.xlabel('Количество итераций', fontsize=14)          
plt.ylabel('log(1 + Loss)', fontsize=14)
plt.title('Loss от количества итераций', fontsize=22)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,6), dpi=75)
for step_alpha in alpha:
    plt.plot(np.cumsum(to_drow[f'{step_alpha}']['time']), to_drow[f'{step_alpha}']['train_acc'], label=f'{step_alpha}')
plt.xlabel('Время', fontsize=14)          
plt.ylabel('Accuracy', fontsize=14)
plt.title('Accuracy от времени на тренировочной выборке', fontsize=22)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,6), dpi=75)
for step_alpha in alpha:
    plt.plot(range(1, 1002), to_drow[f'{step_alpha}']['train_acc'], label=f'{step_alpha}')
plt.xlabel('Количество итераций', fontsize=14)          
plt.ylabel('Accuracy', fontsize=14)
plt.title('Accuracy от количества итераций на тренировочной выборке', fontsize=22)
plt.legend()
plt.show()
```

В целом чем больше step_alpha тем быстрее сходится Градиентный спуск, при этом если взять его слилшком большим графики сильно осциллировать значит мы постоянно 'перепрыгиваем' через локальный минимум.
```python
w_0 = np.zeros(X.shape[1])
sgd = SGDClassifier('binary_logistic',
                    batch_size=10000,
                    step_alpha=0.1,
                    step_beta=0,
                    l2_coef=1e-6)
historysgd = sgd.fit(X,
                  y,
                  w_0,
                  True,
                  X_test=X_test,
                  y_test=y_test)

gdc = GDClassifier('binary_logistic',
                    step_alpha=0.1,
                    step_beta=0,
                    l2_coef=1e-6)
historygdc = gdc.fit(X,
                  y,
                  w_0,
                  True,
                  X_test=X_test,
                  y_test=y_test)
```

```python
plt.figure(figsize=(9,6), dpi=75)
plt.plot(np.cumsum(historygdc['time']), np.log1p(historygdc['func']), label='gdc')
plt.plot(np.cumsum(historysgd['time']), np.log1p(historysgd['func']), label='sgd')
plt.xlabel('Время', fontsize=14)          
plt.ylabel('log(1 + Loss)', fontsize=14)
plt.title('Loss от времени', fontsize=22)
plt.legend()
plt.show()
```

В целом за одно и то же время SGD работает лучше, при этом критерий останова у него срабатывает позже.

Теперь посмотрим на качество в зависимости от min_df.

```python
df_train = pd.read_csv('toxic_train.csv')
y = np.array(df_train['is_toxic'], dtype=int)
y = 2 * y - 1
corpus_train = list(df_train['comment_text'].map(preprocessing))
df_test = pd.read_csv('toxic_test.csv')
y_test = np.array(df_test['is_toxic'], dtype=int)
y_test = y_test * 2 - 1
corpus_test = list(df_test['comment_text'].map(preprocessing))
```

```python
K = [1, 2, 3, 10, 20]
```

```python
to_drow = {}
for k in K:
    vectorizer = CountVectorizer(min_df=k)
    X = vectorizer.fit_transform(corpus_train)
    X_test = vectorizer.transform(corpus_test)
    X = sp.sparse.hstack((X, np.ones(X.shape[0])[:, None]))
    X_test = sp.sparse.hstack((X_test, np.ones(X_test.shape[0])[:, None]))
    X = X.tocsr()
    X_test = X_test.tocsr()
    w_0 = np.zeros(X.shape[1])
    sgd = SGDClassifier('binary_logistic',
                    batch_size=10000,
                    step_alpha=0.1,
                    step_beta=0,
                    l2_coef=1e-6)
    history = sgd.fit(X,
                  y,
                  w_0,
                  True,
                  X_test=X_test,
                  y_test=y_test)
    to_drow[f'{k}'] = history
```

```python
plt.figure(figsize=(9,6), dpi=75)
for k in K:
    plt.plot(np.cumsum(to_drow[f'{k}']['time']), to_drow[f'{k}']['train_acc'], label=f'{k}')
plt.xlabel('Время', fontsize=14)          
plt.ylabel('Accuracy', fontsize=14)
plt.title('Accuracy тренировочный от времени', fontsize=22)
plt.legend()
plt.show()
```

```python
print('Accuracy min_df=20:', max(to_drow['20']['train_acc']))
print('Accuracy min_df=1:', max(to_drow['1']['train_acc']))
```

Разница в  точности при разных min_df практически незаметна, хоть и больше при уменьшении min_df, при этом требуется куда больше времени на обучение, так как увеличивается признаковое пространство.
